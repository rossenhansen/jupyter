```python
class Deck():
    """ A deck of cards  """
    # Colorama 
    from colorama import Fore, Back, Style
    reset = ' \033[0m ' # Reset all colours and backgrounds
    face = '\033[47m' # Set to white
    back = '\033[37;40m  \u265f  ' # Set to black with white pawn symbol
    space = '\033[37m\u265f' # Set to white face pawn symbol
    red = '\033[31m'
    black = '\033[30m'
    yellow = '\033[1;33m'
    cyan = '\033[1;36m'
    # suits
    suits = {
    "H": (red,'\u2665'), # Hearts
    "D": (red,'\u2666'), # Diamonds
    "S": (black,'\u2660'), # Spades
    "C": (black,'\u2663'), # Clubs
    }
    # face cards
    ace = ' A'
    ten = '10'
    jack = ' \u2654J'
    queen = ' \u265BQ'
    king = ' \u265AK'
    joker = '~:)'

    def __init__(self,cards=52):
        self.cards = cards
        self.pack = self.new()

    def card_face(self,rank=0,suit=''):
        # set the card colour
        card_suit = ' '
        if suit in ('H','D','C','S'):
            card_colour = self.suits[suit][0]
            card_suit = self.suits[suit][1]
        elif rank == -1:
            card_colour = self.cyan 
        elif rank == -2:
            card_colour = self.yellow
        else:
            card_colour = self.face # same colour as card face
        # paint the card face
        if rank == 1:
            return ( self.face + self.space + card_colour + self.ace + card_suit + self.reset )
        elif 2 <= rank <= 9:
            card_two_nine = ('  '+str(rank))[-2:]
            return ( self.face + self.space + card_colour + card_two_nine + card_suit + self.reset ) 
        elif rank == 10:
            return ( self.face + self.space + card_colour + self.ten + card_suit + self.reset ) 
        elif rank == 11:
            return ( self.face + card_colour + self.jack + card_suit + self.reset ) 
        elif rank == 12:
            return ( self.face + card_colour + self.queen + card_suit + self.reset ) 
        elif rank == 13:
            return ( self.face + card_colour + self.king + card_suit + self.reset )
        elif -2 <= rank <= -1:
            return  ( self.face + self.space + card_colour + self.joker + self.reset )
        else:
            return self.back

    def card_value(self,rank):
        if rank == 1:
            return 11
        elif rank > 10:
            return 10
        elif rank < 0:
            return 0
        else:
            return rank

    def new(self):
        pack_range = range(1,14)
        if self.cards == 54:
            # add both jokers
            self.pack = [{"rank": rank, "suit": ' ', "value": self.card_value(rank), "face": self.card_face(rank,' ')} for rank in range(-2,0)]
        elif self.cards == 53:
            # add 1 joker
            self.pack = [{"rank": -1, "suit": ' ', "value": self.card_value(-1), "face": self.card_face(-1,' ')}]
        elif self.cards == 43:
            # 500 = 1 joker, aces plus 5 up plus red 4s
            pack_range = range(5,14)
            self.pack = [{"rank": 1, "suit": suit, "value": self.card_value(1), "face": self.card_face(1,suit)} for suit in ('H','D','S','C')] 
            self.pack += [{"rank": 4, "suit": suit, "value": self.card_value(4), "face": self.card_face(4,suit)} for suit in ('H','D')] 
        else:
            self.pack = []
        #build the pack
        self.pack += [{"rank": rank, "suit": suit, "value": self.card_value(rank), "face": self.card_face(rank,suit)} for rank in pack_range for suit in ('H','D','S','C')]
        return self.pack
```
